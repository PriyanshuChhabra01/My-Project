# Exploratory Data Analysis

```python
df.columns
```

**Countplot** 

```python
ax= sns.countplot( data=df, x='Gender', palette=['Pink','skyBlue'] )
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender', y='Amount', data=sales_gen,  palette=['Pink','skyBlue'])
```

**Insights**
: ***From above graphs we can see that most of the buyers are females and even purchasing power of females are greater than men***

**Age**

```python
df.columns
```

```python
ax= sns.countplot(data=df, x='Age Group', hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Total Age Group
sales_Age= df.groupby(["Age Group"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
sns.barplot(data=sales_Age, x="Age Group", y="Amount", palette="Set1")
```

**Inisghts**
: ***From above graphs we can see that most of the Buyers are of age group 26-35 yrs female***

```python
# Total number of orders from top 10 states

sales_state=df.groupby(["State"], as_index=False)["Orders"].sum().sort_values(by="Orders", ascending=False).head(10)
sns.set(rc={"figure.figsize": (16,5)})
sns.barplot(data=sales_state, x="State", y="Orders", palette="Set1")
```
